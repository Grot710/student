This game required me to learn more about the collision detection system, as well as player movement and how to animate characters. I learned how to use the animations so that I could make both the player work by cycling through different frames of the full player spritesheet. Collision detection however, was probably the most important part of the project. To make collision work, I used references from the detection code that was provided from previous games made by students. In the end, me and my group ended up with a level that was good, but still could definitely have been improved. However, overall, I believe that our level was extremely well made for the amount of experience we had and time constraints.
