```python
%%html
<h2>Popcorn Hack 1 Output</h2>
<div id="output1"></div>

<script>
  (function() {
    class Dice {
      constructor(sides) {
        this.sides = sides;
      }

      roll() {
        return Math.floor(Math.random() * this.sides) + 1;
      }
    }

    const newDice = new Dice(6);
    const output = document.getElementById("output1");
    
    output.innerHTML += "Roll: " + newDice.roll() + "<br>";
  })();
</script>
```

```python
%%html
<h2>üêæ Popcorn Hack 2 - Pet Simulator üêæ</h2>
<div id="petOutput"></div>
<button onclick="myPet.feed()">Feed</button>
<button onclick="myPet.play()">Play</button>
<button onclick="myPet.describe()">Check Status</button>

<script>
  class Pet {
    constructor(name, hunger, energy, happiness) {
      this.name = name;
      this.hunger = Number(hunger);
      this.energy = Number(energy);
      this.happiness = Number(happiness);
    }

    feed() {
      this.hunger = Math.max(this.hunger - 3, 0); // prevent negative hunger
      this.energy = Math.min(this.energy + 2, 10);
      this.happiness += 1;
      this.describe("Your pet has been fed!");
    }

    play() {
      this.energy = Math.max(this.energy - 2, 0);
      this.hunger = Math.min(this.hunger + 2, 10);
      this.happiness += 3;
      this.describe("You played with your pet!");
    }

    describe(message = "") {
      const output = document.getElementById("petOutput");
      output.innerHTML = `
        <strong>${message}</strong><br>
        <strong>Name:</strong> ${this.name}<br>
        <strong>Hunger:</strong> ${this.hunger}<br>
        <strong>Energy:</strong> ${this.energy}<br>
        <strong>Happiness:</strong> ${this.happiness}<br>
      `;
    }
  }

  // Create a pet
  const myPet = new Pet("Baller", 5, 5, 5);
  myPet.describe("Meet your pet!");
</script>
```
