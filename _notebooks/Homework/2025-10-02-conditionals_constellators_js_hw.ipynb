```python
%%html

<div id="outputHW">Results will appear here:<br></div>
<button onclick="checkTemp()">Check Temperature</button>
<button onclick="checkScore()">Check Score</button>
<button onclick="randomNumb()">Random Number 1-10</button>
<button onclick="checkDay()">Check Day</button>
<button onclick="checkLogin()">Check Login</button>
<button onclick="checkWeather()">Check Weather</button>

<script>
(() => {

  const output = document.getElementById("outputHW");
  output.innerText += "Code, code, code!\n";

  // 1) Temperature check
  window.checkTemp = function () {
    let temperature = Number(prompt("Enter a temperature:"));
    if (temperature > 80) {
      output.innerText += "It's hot outside!\n";
    } else {
      output.innerText += "It's nice outside.\n";
    }
  };

  // 2 & 3) Score check with prompt
  window.checkScore = function () {
    let score = Number(prompt("Enter your score (1-100):"));
    if (score >= 90) {
      output.innerText += "A\n";
    } else if (score >= 80) {
      output.innerText += "B\n";
    } else if (score >= 70) {
      output.innerText += "C\n";
    } else {
      output.innerText += "Needs improvement.\n";
    }
  };

  // 4 & 5) Random number between 1-10, check if even or odd
  window.randomNumb = function () {
    let number = Math.floor(Math.random() * 10) + 1;
    if (number % 2 === 0) {
      output.innerText += `The number ${number} is even.\n`;
    } else {
      output.innerText += `The number ${number} is odd.\n`;
    }
  };

  // 6 & 7) Day check with switch, plus weekend/weekday
  window.checkDay = function () {
    let day = prompt("Enter a day of the week (e.g., Monday):");
    if (!day) return; // User cancelled prompt
    day = day.trim().toLowerCase(); // Normalize input

    // Message for specific day
    switch (day) {
      case "monday":
        output.innerText += "Ugh, it's Monday...\n";
        break;
      case "tuesday":
        output.innerText += "It's Tuesday!\n";
        break;
      case "wednesday":
        output.innerText += "Late start\n";
        break;
      case "thursday":
        output.innerText += "Almost Friday!\n";
        break;
      case "friday":
        output.innerText += "TGIF\n";
        break;
      case "saturday":
        output.innerText += "Sat!\n";
        break;
      case "sunday":
        output.innerText += "Sun.\n";
        break;
      default:
        output.innerText += "Not a valid day.\n";
        return;
    }

    // Weekend or weekday check
    if (day === "saturday" || day === "sunday") {
      output.innerText += "Weekend!\n";
    } else {
      output.innerText += "Weekday!\n";
    }
  };

  // 8) Boolean check for login
  window.checkLogin = function () {
    let response = prompt("Are you logged in? (yes/no)").toLowerCase();
    let loggedIn = response === "yes" || response === "y";
    if (loggedIn) {
      output.innerText += "Welcome back!\n";
    } else {
      output.innerText += "Please log in.\n";
    }
  };

  // 9) Weather check
  window.checkWeather = function () {
    let weather = prompt("Is it raining? (yes/no)").toLowerCase();
    if (weather === "yes" || weather === "y") {
      output.innerText += "It's raining.\n";
    } else {
      output.innerText += "Go outside, it's a nice day today!\n";
    }
  };

})();
</script>
```
